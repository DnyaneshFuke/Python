```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
%matplotlib inline
plt.xlabel("Years")
plt.ylabel("Income")
plt.scatter(df.year,df.income ,color="red" ,marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']], df['income'])
```

```python
reg.predict([[1969]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d=pd.read_csv("R_Y.csv")
d[:5]
```

```python
p=reg.predict(d)
d['Income']=p
```
